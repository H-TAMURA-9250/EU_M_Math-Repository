機械学習や統計学の重要キーワードである。これらはデータから予測モデルを作ったり、データを分類したりする際の「目的」「手法」「評価基準」に分類される。それぞれの役割、意味、活用場面について解説する。

1. 基本的な概念・枠組み
教師あり学習 (Supervised Learning)
意味・役割: 正解（ラベル）があらかじめ与えられたデータを使ってモデルを学習させる手法である。

活用場面: 過去の売上データ（正解あり）から未来の売上を予測する場合や、手書き文字の画像（正解ラベルあり）から文字を認識させる場合に用いる。

回帰 (Regression)
意味・役割: 連続する数値（量）を予測するタスクである。教師あり学習の一種。

活用場面: 明日の気温、株価、マンションの価格、製品の需要予測など。

分類 (Classification)
意味・役割: データをあらかじめ決められたカテゴリ（クラス）に振り分けるタスクである。これも教師あり学習の一種。

活用場面: メールのスパム判定（スパムか否か）、画像の物体認識（犬か猫か）、顧客の離脱予測（離脱するか継続するか）。

2. 回帰・分類の具体的な手法
重回帰分析 (Multiple Linear Regression)
意味・役割: 1つの数値（目的変数）を、複数の要因（説明変数）から直線的な式で予測する回帰手法。

活用場面: 家賃を予測するために「駅からの距離」「築年数」「広さ」の3つの要素を使って分析する場合。

ロジスティック回帰分析 (Logistic Regression)
意味・役割: 名前に「回帰」と付くが、実際には2値分類（Yes/No）を行う確率を予測する手法。結果が0から1の間の確率で出力される。

活用場面: ある病気に「罹患しているか・していないか」の確率予測、バナー広告が「クリックされるか・されないか」の予測。

k-NN法（k近傍法 / k-Nearest Neighbors）
意味・役割: 新しいデータの近くにある k 個の既存データを参照し、多数決で分類（または平均値で回帰）を決めるシンプルな手法。

活用場面: 推薦システム（似たような購買傾向を持つ他のユーザー k 人が買っている商品を薦める）など。

SVM（サポートベクターマシン / Support Vector Machine）
意味・役割: データ群を境界線（マージン）で2つに切り分ける際、境界とデータの距離が最も遠くなるような最適な線（面）を見つける手法。非常に高い識別性能を持つ。

活用場面: 文字認識、音声認識、遺伝子データの分類など、境界線が複雑になりがちな高次元データの分類。

3. 決定木とその評価基準
決定木 (Decision Tree)
意味・役割: データを「Yes/No」の条件分岐（樹形図のような構造）で繰り返し分割し、分類や回帰を行う手法。人間が見て理由を理解しやすい（解釈性が高い）特徴がある。

活用場面: 銀行の融資審査（年収＞○○万か？ → 勤続年数＞○年か？ → 融資OK/NG）の自動化。

エントロピー (Entropy)
意味・役割: データ全体の「乱雑さ（ごちゃ混ぜ度）」を表す指標。すべてのデータが1つのクラスに揃っていればエントロピーは 0（綺麗に分かれている状態）になり、半々だと最大になる。

活用場面: 決定木において、データをどの条件で分割すれば「より綺麗に（純度高く）グループ分けできるか」を測る基準として内部で計算される。

情報利得 (Information Gain)
意味・役割: データをある条件で分割した際、「どれだけエントロピー（乱雑さ）を減少させられたか」を表す指標。

活用場面: 決定木を作る際、「どの質問（条件）を最初に持ってくるのが最も効率的か」を判断するために、情報利得が最大になる条件を探索する場面で使う。

4. 過学習を防ぐアプローチ
正則化 (Regularization)
意味・役割: モデルが学習データに適合しすぎて、未知のデータに対応できなくなる現象（過学習 / オーバーフィッティング）を防ぐ技術。数式に「複雑さへのペナルティ」を加えることで、モデルをシンプルに保つ。

活用場面: データ数が少ないのに対して特徴量（変数）が多すぎ、過学習を起こしそうな場面で広く使われる。

リッジ回帰 (Ridge Regression)
意味・役割: 正則化の一種（L2正則化）。係数（重み）の2乗の和をペナルティとして科すことで、各変数の影響力を全体的に小さく（なだらかに）抑える回帰手法。

活用場面: 多くの変数同士に強い相関（多重共線性）があり、通常の重回帰分析では予測が不安定になる場面。

ラッソ回帰 (Lasso Regression)
意味・役割: 正則化の一種（L1正則化）。係数の絶対値の和をペナルティとして科す。特徴として、不要な変数の係数を完全に「0」にする性質がある。

活用場面: 数百個ある特徴量の中から、本当に予測に必要な少数の変数だけを自動的に選び出したい（特徴量選択）場面。

5. 機械学習の前提となる定理
ノーフリーランチ定理 (No Free Lunch Theorem)
意味・役割: 「あらゆる問題に対して、常に他の手法より優れている万能な機械学習アルゴリズムは存在しない」という数学的定理。

活用場面: 手法を選ぶ際、「最新のディープラーニングやSVMを使えばどんなデータでも一番上手くいく」と過信するのではなく、扱うデータの性質や目的に応じて複数の手法（K-meansやロジスティック回帰など）を試し、比較検討すべきというデータサイエンティストの心得・指針として引用される。
