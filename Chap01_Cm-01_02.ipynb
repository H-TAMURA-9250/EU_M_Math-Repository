```python
#Chap01_Cm-01
def is_prime(n):
    if n<= 1:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True


def list_primes(start, end):
    primes = []
    for num in range(start, end + 1):
        if is_prime(num):
            primes.append(num)
    return primes

start = int(input("begin_num:"))
end = int(input("end_num:"))

primes = list_primes(start, end)
print(f"from {start} to {end} range is that:{primes}")

```

```python
#Chap01_Cm-02
def get_primes(n):
    if n < 2:
        return []

    # 0からnまでの真偽値リストを作成（初期値はすべてTrue）
    is_prime = [True] * (n + 1)
    is_prime[0] = is_prime[1] = False  # 0と1は素数ではない

    # エラトステネスの篩を実行
    for i in range(2, int(n**0.5) + 1):
        if is_prime[i]:
            # iが素数なら、その倍数をすべてFalseにする
            for j in range(i * i, n + 1, i):
                is_prime[j] = False

    # Trueのまま残ったインデックス（数値）をリストにして返す
    return [i for i, prime in enumerate(is_prime) if prime]

# 実行例
N = int(input("Nを入力してください: "))
primes = get_primes(N)
print(f"{N}までの素数: {primes}")
print(f"素数の個数: {len(primes)}")
```
