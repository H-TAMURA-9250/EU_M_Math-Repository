```python
# Chap02_Ex-1

def sum_natural_numbers_loop(n):
    """ループを使って1からnまでの自然数の和を計算する"""
    total = 0
    for i in range(1, n + 1):
        total += i
    return total

def sum_natural_numbers_formula(n):
    """公式を使って1からnまでの自然数の和を計算する"""
    return n * (n + 1) // 2

# 例として、1から10までの自然数の和を計算
n_example = 10

sum_loop = sum_natural_numbers_loop(n_example)
sum_formula = sum_natural_numbers_formula(n_example)

print(f"1から{n_example}までの自然数の和 (ループ): {sum_loop}")
print(f"1から{n_example}までの自然数の和 (公式): {sum_formula}")

# 別の例
n_another_example = 100
sum_loop_100 = sum_natural_numbers_loop(n_another_example)
print(f"1から{n_another_example}までの自然数の和 (ループ): {sum_loop_100}")
```

```python
#Chap02_Ex-9
from pandas import Series, DataFrame
import pandas as pd

# attri_data_framel の定義 (セルの独立性を確保するため再定義)
attri_datal = {'ID':['1','2','3','4','5'],
               'Sex':['F','F','M','M','F'],
               'Money':['1000','2000','500','300','700'],
               'Name':['Saito','Horie','Kondo','Kawada','Matsubara'],}
attri_data_framel = DataFrame(attri_datal)
attri_data_framel['Money'] = attri_data_framel['Money'].astype(int) # 'Money'列を数値型に変換

# attri_data_frame2 の定義
attri_data2 = {'ID':['3','4','7'],
               'Math':[60,30,40],
               'English':[80,20,30]}
attri_data_frame2 = DataFrame(attri_data2)

# データをマージ（IDをキーとして内部結合）
merged_df = pd.merge(attri_data_framel, attri_data_frame2, on='ID', how='inner')

# 'Money', 'Math', 'English' の平均を計算
average_values = merged_df[['Money', 'Math', 'English']].mean()

# 結果を表示
print('Money, Math, English の平均:')
display(average_values)
```

```python
#Chap02_Ex-12
import numpy as np
import matplotlib.pyplot as plt

def plot_uniform_histograms(N):
    """N個の一様乱数2組のヒストグラムを描画する"""
    # 0から1の一様乱数をN個、2組発生
    data1 = np.random.uniform(0.0, 1.0, N)
    data2 = np.random.uniform(0.0, 1.0, N)

    # plt.subplotを使ってヒストグラムを並べて表示
    fig, axes = plt.subplots(1, 2, figsize=(12, 5))
    fig.suptitle(f'Uniform Random Numbers (N={N})', fontsize=16)

    # 1つ目のヒストグラム
    axes[0].hist(data1, bins=20, edgecolor='black', alpha=0.7)
    axes[0].set_title(f'Histogram of Uniform Random Numbers 1 (N={N})')
    axes[0].set_xlabel('Value')
    axes[0].set_ylabel('Frequency')
    axes[0].set_xlim(0, 1) # 0から1の範囲に固定

    # 2つ目のヒストグラム
    axes[1].hist(data2, bins=20, edgecolor='black', alpha=0.7, color='orange')
    axes[1].set_title(f'Histogram of Uniform Random Numbers 2 (N={N})')
    axes[1].set_xlabel('Value')
    axes[1].set_ylabel('Frequency')
    axes[1].set_xlim(0, 1) # 0から1の範囲に固定

    plt.tight_layout(rect=[0, 0.03, 1, 0.95]) # タイトルとサブプロットの重なりを避ける
    plt.show()

# 100個の乱数で実施
plot_uniform_histograms(100)
```

```python
# 1000個の乱数で実施
plot_uniform_histograms(1000)
```

```python
# 10000個の乱数で実施
plot_uniform_histograms(10000)
```

```python
#考察
#N=100 (100個の乱数): ヒストグラムの各ビン（区間）の高さがかなりばらついており、一様分布であることが分かりにくい,サンプルサイズが小さいため、ランダムなばらつきが大きく影響している。
#N=1000 (1000個の乱数): N=100の場合と比べて、各ビンの高さのばらつきが減少し、より平坦な（一様分布に近い）形状になってきていることが視覚的に確認できる。サンプルサイズが増えたことで、ランダムな変動が平均化されたためです。
#N=10000 (10000個の乱数): さらにサンプルサイズが増えることで、ヒストグラムは非常に平坦になり、各ビンにほぼ均等な数のデータが含まれていることが明確に分かる。これは、大数の法則が働いているため。試行回数が増えるほど、経験的な平均が理論的な値に近づいていく。
#結論として、乱数の個数（サンプルサイズ）が増えるほど、その乱数のヒストグラムは理論的な一様分布の形状（平坦な長方形）に近づいていくことが分かる。
```
