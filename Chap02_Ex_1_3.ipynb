```python
#EX-01
import numpy as np

# Generate the first set of 10,000 uniform random numbers in [0, 1] using np.random.uniform
random_numbers_set1 = np.random.uniform(low=0.0, high=1.0, size=10000)

# Generate the second set of 10,000 uniform random numbers in [0, 1] using np.random.uniform
random_numbers_set2 = np.random.uniform(low=0.0, high=1.0, size=10000)

print("Shape of the first set:", random_numbers_set1.shape)
print("First 5 elements of the first set:", random_numbers_set1[:5])
print("\nShape of the second set:", random_numbers_set2.shape)
print("First 5 elements of the second set:", random_numbers_set2[:5])
```

```python
#EX-02
import matplotlib.pyplot as plt

# Combine the two sets of random numbers into (x, y) coordinates
x_coords = random_numbers_set1
y_coords = random_numbers_set2

# Calculate the Euclidean distance from the origin for each point
distances = np.hypot(x_coords, y_coords)

# Determine which points fall inside the unit circle (radius 1)
inside_circle = distances < 1

# Count the number of points inside the circle
num_points_inside_circle = np.sum(inside_circle)
total_points = len(x_coords)

# Estimate Pi
pi_estimate = 4 * (num_points_inside_circle / total_points)

print(f"Total points generated: {total_points}")
print(f"Points inside the unit circle: {num_points_inside_circle}")
print(f"Estimated value of Pi: {pi_estimate:.6f}")
```

```python
#EX-02
# Create a scatter plot
plt.figure(figsize=(8, 8))

# Plot points inside the circle in blue
plt.scatter(x_coords[inside_circle], y_coords[inside_circle], color='blue', s=1, label='Inside Circle')

# Plot points outside the circle in red
plt.scatter(x_coords[~inside_circle], y_coords[~inside_circle], color='red', s=1, label='Outside Circle')

# Draw the quarter circle boundary for reference
theta = np.linspace(0, np.pi/2, 100)
plt.plot(np.cos(theta), np.sin(theta), color='green', linewidth=2, label='Unit Quarter Circle Boundary')

plt.title('Monte Carlo Pi Estimation: Points within Unit Square')
plt.xlabel('X-coordinate')
plt.ylabel('Y-coordinate')
plt.xlim(0, 1)
plt.ylim(0, 1)
plt.gca().set_aspect('equal', adjustable='box') # Ensure equal scaling for x and y axes
plt.legend()
plt.grid(True)
plt.show()
```
