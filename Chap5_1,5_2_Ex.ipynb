```python
#Chap05_Ex-02
import numpy as np
import numpy.random as random
import scipy as sp

import matplotlib.pyplot as plt
import matplotlib as mpl
%matplotlib inline

# Set print options for precision to 3 decimal places
np.set_printoptions(precision=3)

sample_names = np.array(['a' , 'b' ,'c' , 'd ', 'a'])
random.seed(0)
data = random.randn(5,5)

# Apply boolean indexing to filter rows of the 'data' array for all samples except 'c'
data = data[sample_names != 'c']

print(sample_names)
print(data)
```

```python
#Chap05_Ex-02
sample_list = [1, 2, 3, 4, 5]
print(f"元のリスト: {sample_list}")

# リストをNumPy配列に変換
sample_array = np.array(sample_list)

# ブロードキャスト機能を使って各要素に3を加える
result_array = sample_array + 3

print(f"3を加えたNumPy配列: {result_array}")

# 必要であれば、再度リストに変換することも可能です
result_list = result_array.tolist()
print(f"3を加えたリスト: {result_list}")
```
