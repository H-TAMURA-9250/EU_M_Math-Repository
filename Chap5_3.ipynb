```python
#Ex-18
from scipy.optimize import fsolve

def g(x):
    return x**3 - 2*x**2 - 11*x + 12

# Initial guess for the root. This function has multiple roots, so the guess matters.
x0_g = 1 # Let's try an initial guess of 1

root_g = fsolve(g, x0_g)
print(f"The root of f(x) = x^3 - 2x^2 - 11x + 12 with an initial guess of {x0_g} is: {root_g[0]:.2f}")

x0_g_2 = -3 # Another common root is -3
root_g_2 = fsolve(g, x0_g_2)
print(f"The root of f(x) = x^3 - 2x^2 - 11x + 12 with an initial guess of {x0_g_2} is: {root_g_2[0]:.2f}")

x0_g_3 = 4 # Another common root is 4
root_g_3 = fsolve(g, x0_g_3)
print(f"The root of f(x) = x^3 - 2x^2 - 11x + 12 with an initial guess of {x0_g_3} is: {root_g_3[0]:.2f}")
```

```python
#Cm-1
import numpy as np
import scipy.linalg as spla

A = np.array([[5,1,0,1],[1,9,-5,7],[0,-5,8,-3],[1,7,-3,10]])
b = np.array([2,10,5,10])

print('Matrix A =\n', A)
print('\nVector b =\n', b)

# Perform Cholesky decomposition
# L is the lower triangular Cholesky factor
L = spla.cholesky(A, lower=True)
print('\nCholesky factor L =\n', L)

# Solve L @ y = b for y
y = spla.solve(L, b)
print('\nIntermediate vector y =\n', y)

# Solve L.T @ x = y for x
x_cholesky = spla.solve(L.T, y)
print('\nSolution x =\n', x_cholesky)

# Verify the solution
print('\nVerification: A @ x =\n', A @ x_cholesky)
```

```python
#Cm-2
from sympy import integrate, Symbol, sqrt, pi

x, y, u, v = Symbol('x'), Symbol('y'), Symbol('u'), Symbol('v')

# Define the integrand in terms of u
integrand_u = 1 / (sqrt(u) * (1 + u)**2)

# Calculate the inner integral with respect to u
# from v to 1
inner_integral_result = integrate(integrand_u, (u, v, 1))
print(f"Result of inner integral (with respect to u): {inner_integral_result}")

# Calculate the outer integral with respect to v
# from 0 to 1
total_integral_result = integrate(inner_integral_result, (v, 0, 1))
print(f"\nFinal integral value: {total_integral_result}")

# Evaluate numerically for a clearer understanding of the value
print(f"Numerical approximation: {total_integral_result.evalf()}")
```

```python
#Cm-3
from scipy.optimize import minimize

# Define the objective function
def objective_function(x):
    return x**2 + 1

# Define the bounds for x: x > -1
# (lower_bound, upper_bound) - None means no upper bound
bounds = [(-1, None)]

# Initial guess for x
x0 = 0.5 # A value satisfying the constraint and close to the expected minimum

# Perform the optimization
result = minimize(objective_function, x0, bounds=bounds, method='L-BFGS-B')

print(f"Optimization result:\n{result}")
print(f"\nMinimum value of x: {result.x[0]:.4f}")
print(f"Minimum function value: {result.fun:.4f}")
```
