## ベイズの定理による病気Xの確率計算

与えられた問題は、ある検査で陽性反応が出た場合に、実際に病気Xに罹患している確率を求めるものです。これはベイズの定理を用いて計算できます。

**ベイズの定理:**
$P(A|B) = \frac{P(B|A) \cdot P(A)}{P(B)}$

ここで、
*   $A$: 病気Xに罹患している事象
*   $B$: 検査で陽性反応が出る事象

与えられた情報:
*   $P(A)$ (病気Xの有病率) = 0.1% = 0.001
*   $P(B|A)$ (病気Xの人が陽性反応を示す確率、感度) = 99% = 0.99
*   $P(B|\neg A)$ (病気Xでない人が陽性反応を示す確率、偽陽性率) = 3% = 0.03

まず、$P(B)$ (陽性反応が出る確率) を計算する必要があります。
$P(B) = P(B|A) \cdot P(A) + P(B|\neg A) \cdot P(\neg A)$
$P(\neg A) = 1 - P(A) = 1 - 0.001 = 0.999$

```python
# 全体のくじの数
total_tickets = 1000
# 当たりのくじの数
winning_tickets = 100

# Aが当たる確率
prob_a_wins = winning_tickets / total_tickets
print(f"Aが当たる確率: {prob_a_wins:.3f}")

# Aが当たった後、残りのくじの数と当たりのくじの数
remaining_total_tickets = total_tickets - 1
remaining_winning_tickets = winning_tickets - 1

# Bが当たる確率 (Aが当たった後)
prob_b_wins_given_a_wins = remaining_winning_tickets / remaining_total_tickets
print(f"Aが当たった後、Bが当たる確率: {prob_b_wins_given_a_wins:.3f}")

# AとBがともに当たる確率
prob_both_win = prob_a_wins * prob_b_wins_given_a_wins
print(f"AとBがともに当たる確率: {prob_both_win:.3f}")
```

```python
import numpy as np

# Given probabilities
# P(A): 有病確率 (Prevalence of disease X)
prob_disease = 0.001  # 0.1%

# P(B|A): 病気の人が陽性反応を示す確率 (Sensitivity)
prob_positive_given_disease = 0.99  # 99%

# P(B|~A): 病気でない人が陽性反応を示す確率 (False positive rate)
prob_positive_given_no_disease = 0.03  # 3%

# P(~A): 病気でない確率
prob_no_disease = 1 - prob_disease

# P(B): 陽性反応が出る確率 (Probability of a positive test result)
# P(B) = P(B|A) * P(A) + P(B|~A) * P(~A)
prob_positive_test = (prob_positive_given_disease * prob_disease) + \
                     (prob_positive_given_no_disease * prob_no_disease)

# P(A|B): 陽性反応が出たときに病気Xである確率 (Posterior probability)
# P(A|B) = [P(B|A) * P(A)] / P(B)
prob_disease_given_positive = (prob_positive_given_disease * prob_disease) / prob_positive_test

print(f"病気Xの有病率: {prob_disease:.3%}")
print(f"病気Xの人が陽性反応を示す確率 (感度): {prob_positive_given_disease:.0%}")
print(f"病気Xでない人が陽性反応を示す確率 (偽陽性率): {prob_positive_given_no_disease:.0%}")
print(f"\n検査で陽性反応が出た人が病気Xである確率: {prob_disease_given_positive:.2%}")
```

## 標本平均の不偏性の証明 (Demonstration of Unbiasedness of Sample Mean)

平均$\mu$、分散$\sigma^2$の正規母集団から抽出された大きさ$n$の標本（$X_1, X_2, ..., X_n$）から作られる標本平均$\bar{X}$は、母平均$\mu$の不偏推定量であることを示します。

### 不偏推定量とは

推定量$\hat{\theta}$がパラメータ$\theta$の不偏推定量であるとは、その期待値が真のパラメータに等しいことを意味します。数式で表すと $E[\hat{\theta}] = \theta$ となります。

### 標本平均$\bar{X}$の期待値

標本平均$\bar{X}$は次のように定義されます。

$\bar{X} = \frac{1}{n} \sum_{i=1}^{n} X_i$

この標本平均の期待値を計算します。

$E[\bar{X}] = E\left[\frac{1}{n} \sum_{i=1}^{n} X_i\right]$

期待値の線形性（$E[aY + bZ] = aE[Y] + bE[Z]$）を利用すると、定数$1/n$を期待値の外に出すことができます。

$E[\bar{X}] = \frac{1}{n} E\left[\sum_{i=1}^{n} X_i\right]$

さらに、和の期待値は期待値の和に等しいという性質を利用します。

$E[\bar{X}] = \frac{1}{n} \sum_{i=1}^{n} E[X_i]$

ここで、$X_i$は全て同じ母集団（平均$\mu$）から抽出されたものであるため、各$E[X_i]$は母平均$\mu$に等しくなります。

$E[X_i] = \mu \quad (\text{for all } i=1, ..., n)$

これを上記の式に代入すると、

$E[\bar{X}] = \frac{1}{n} \sum_{i=1}^{n} \mu$

和は$n$個の$\mu$の合計となるため、

$E[\bar{X}] = \frac{1}{n} (n\mu)$

したがって、

$E[\bar{X}] = \mu$

この結果から、標本平均$\bar{X}$の期待値が母平均$\mu$に等しいことが示されました。これは、標本平均が母平均の不偏推定量であることを意味します。

### 数値シミュレーションによる確認

この理論的な性質を、実際にシミュレーションで確認してみましょう。平均0、標準偏差1の正規分布から多数の標本を抽出し、それぞれの標本平均を計算し、その平均が母平均に近づくことを示します。

```python
import numpy as np
import matplotlib.pyplot as plt

# 母集団のパラメータ
population_mean = 0      # 母平均 μ
population_std = 1       # 母標準偏差 σ

# シミュレーションのパラメータ
sample_size = 100        # 各標本の大きさ n
num_simulations = 100000 # シミュレーションの繰り返し回数

# 標本平均を格納するリスト
sample_means_for_unbiasedness = []

# シミュレーションの実行
for _ in range(num_simulations):
    # 母集団からn個の標本を抽出
    samples = np.random.normal(loc=population_mean, scale=population_std, size=sample_size)
    # 標本平均を計算してリストに追加
    sample_means_for_unbiasedness.append(np.mean(samples))

# 標本平均の平均を計算
average_of_sample_means = np.mean(sample_means_for_unbiasedness)

print(f"真の母平均 (μ): {population_mean}")
print(f"標本平均の平均 (E[X̄]の近似値): {average_of_sample_means:.4f}")

# 結果をヒストグラムで可視化 (任意)
plt.figure(figsize=(10, 6))
plt.hist(sample_means_for_unbiasedness, bins=50, density=True, alpha=0.7, color='steelblue', edgecolor='black')
plt.axvline(population_mean, color='red', linestyle='dashed', linewidth=2, label=f'母平均 (μ = {population_mean})')
plt.axvline(average_of_sample_means, color='green', linestyle='dotted', linewidth=2, label=f'標本平均の平均 ({average_of_sample_means:.4f})')
plt.title('標本平均の分布と母平均 (不偏性の確認)')
plt.xlabel('標本平均')
plt.ylabel('頻度 (密度)')
plt.legend()
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
import numpy as np

# 真のパラメータ λ
true_lambda = 0.5

# サンプルサイズ
sample_size_exp = 1000

# 指数分布からの乱数生成
# np.random.exponential は scale = 1/lambda をパラメータとする
samples_exp = np.random.exponential(scale=1/true_lambda, size=sample_size_exp)

# 標本平均を計算
sample_mean_exp = np.mean(samples_exp)

# 最尤推定量 λ_hat を計算
mle_lambda = 1 / sample_mean_exp

print(f"真のパラメータ λ: {true_lambda}")
print(f"生成された標本サイズ: {sample_size_exp}")
print(f"標本平均 (X̄): {sample_mean_exp:.4f}")
print(f"最尤推定量 λ̂ (1/X̄): {mle_lambda:.4f}")

# 結果をMarkdownセルとして出力するためにdisplayを使用
from IPython.display import Markdown, display
display(Markdown(
    f"真のパラメータ $\lambda = {true_lambda}$ に対し、生成された標本からの最尤推定量 $\hat{{\lambda}}$ は **{mle_lambda:.4f}** となりました。\n"
    f"これは、標本平均が真の期待値 ($1/\lambda$) に近づくにつれて、$\hat{{\lambda}}$ が真の $\lambda$ に近づくことを示しています。"
))
```

## Identifying Influential Factors on G3 Grades

To understand which factors influence student performance, we will prepare our merged dataset for predictive modeling. This involves encoding categorical features and then using a regression model to assess the importance of different student attributes on their final grades (G3) in both mathematics and Portuguese.

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler

# Prepare the features (X) and target variables (y)
# We will use all common columns as features, and G3_mat and G3_por as targets

# Drop the G1 and G2 grades from features as G3 is the final grade
# Also drop the G3 grades for each subject from the feature set itself
feature_cols = [col for col in common_cols if col not in ['G1', 'G2', 'G3']] # This will already be handled by common_cols selection above, but explicitly checking for G1/G2 for clarity

X = df_merged[common_cols].copy()
y_mat = df_merged['G3_mat']
y_por = df_merged['G3_por']

# Identify categorical columns for one-hot encoding
categorical_cols = X.select_dtypes(include=['object']).columns

# Apply one-hot encoding to categorical features
X_encoded = pd.get_dummies(X, columns=categorical_cols, drop_first=True)

print("Shape of encoded features X:", X_encoded.shape)
display(X_encoded.head())
```

### Linear Regression Model for Mathematics G3 Grades

```python
# Split data for Mathematics G3
X_train_mat, X_test_mat, y_train_mat, y_test_mat = train_test_split(X_encoded, y_mat, test_size=0.2, random_state=42)

# Scale numerical features (optional but good practice for regression)
scaler = StandardScaler()
X_train_mat_scaled = scaler.fit_transform(X_train_mat)
X_test_mat_scaled = scaler.transform(X_test_mat)

# Initialize and train the Linear Regression model for Math
model_mat = LinearRegression()
model_mat.fit(X_train_mat_scaled, y_train_mat)

# Make predictions and evaluate the model
y_pred_mat = model_mat.predict(X_test_mat_scaled)
mse_mat = mean_squared_error(y_test_mat, y_pred_mat)
r2_mat = r2_score(y_test_mat, y_pred_mat)

print(f"Mathematics G3 - Mean Squared Error: {mse_mat:.2f}")
print(f"Mathematics G3 - R-squared: {r2_mat:.2f}")

# Get feature importances (coefficients)
feature_importance_mat = pd.DataFrame({
    'Feature': X_encoded.columns,
    'Coefficient': model_mat.coef_
}).sort_values(by='Coefficient', ascending=False)

print("\nTop 10 Most Influential Factors for Mathematics G3 (Positive Coefficients):")
display(feature_importance_mat.head(10))
print("\nTop 10 Most Influential Factors for Mathematics G3 (Negative Coefficients):")
display(feature_importance_mat.tail(10))
```

```python
# Split data for Portuguese G3
X_train_por, X_test_por, y_train_por, y_test_por = train_test_split(X_encoded, y_por, test_size=0.2, random_state=42)

# Scale numerical features
X_train_por_scaled = scaler.fit_transform(X_train_por)
X_test_por_scaled = scaler.transform(X_test_por)

# Initialize and train the Linear Regression model for Portuguese
model_por = LinearRegression()
model_por.fit(X_train_por_scaled, y_train_por)

# Make predictions and evaluate the model
y_pred_por = model_por.predict(X_test_por_scaled)
mse_por = mean_squared_error(y_test_por, y_pred_por)
r2_por = r2_score(y_test_por, y_pred_por)

print(f"Portuguese G3 - Mean Squared Error: {mse_por:.2f}")
print(f"Portuguese G3 - R-squared: {r2_por:.2f}")

# Get feature importances (coefficients)
feature_importance_por = pd.DataFrame({
    'Feature': X_encoded.columns,
    'Coefficient': model_por.coef_
}).sort_values(by='Coefficient', ascending=False)

print("\nTop 10 Most Influential Factors for Portuguese G3 (Positive Coefficients):")
display(feature_importance_por.head(10))
print("\nTop 10 Most Influential Factors for Portuguese G3 (Negative Coefficients):")
display(feature_importance_por.tail(10))
```
